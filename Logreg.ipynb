```python
from sklearn import datasets
X, y = datasets.make_classification(n_samples=100, n_features=3, n_informative=3, n_redundant=0, random_state=42)
import numpy as np
```

```python
def fn(X, Y):
    w = np.zeros(X.shape[1] + 1)
    XX = np.ones([X.shape[0], X.shape[1] + 1])
    for i in range(X.shape[0]):
        for j in range(X.shape[1]):
            XX[i][j] = X[i][j]
    for q in range(1000): 
        u = 0
        delt = []
        y = np.zeros(Y.shape)
        for i in range(len(Y)):
            for j in range(XX.shape[1]):
                y[i] += XX[i][j] * w[j] 
        y_exp = np.exp(np.array(-y))
        sigm = 1/(1+y_exp[j])
        for i in range(len(y)):
            y[i] = 1/(1 + y_exp[i])
        for i in range(len(y)):
            u += Y[i] * np.log(1/(1+y_exp[i])) + (1-Y[i]) * np.log(1-1/(1+y_exp[i]))
        u /= len(Y)
        print(u)
        for i in range(XX.shape[1]):
            o = 0
            for j in range(len(Y)):
                o += Y[j] * sigm * y_exp[j] * XX[j][i] - (1-Y[j]) * 1/(1-sigm) * sigm **2 * y_exp[j] * XX[j][i]
            o = o / len(Y)
            delt.append(o)
        for t in range(len(w)):
            w[t] = w[t] + 0.01 * delt[t]
    return w
```
